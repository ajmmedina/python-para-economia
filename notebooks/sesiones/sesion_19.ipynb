## 1. Carga de datos financieros

Ejemplo:

```python
import pandas as pd

datos = pd.read_csv('datos_financieros.csv')
print(datos.head())
```

## 2. Cálculo de métricas de riesgo

Ejemplo:

```python
import numpy as np

retornos = datos['retorno'].values
VaR_95 = np.percentile(retornos, 5)
print(f"VaR al 95%: {VaR_95:.2f}")
```

## 3. Simulaciones Monte Carlo

Ejemplo:

```python
simulaciones = np.random.normal(np.mean(retornos), np.std(retornos), 10000)
VaR_sim = np.percentile(simulaciones, 5)
print(f"VaR simulado al 95%: {VaR_sim:.2f}")
```
