## 1. ¿Qué es una base de datos?

Una base de datos es un sistema organizado para almacenar, gestionar y recuperar información de manera eficiente.

#### Ejemplo aplicado a economía: 

Una base de datos que almacena información de clientes, productos y ventas.

## 2. Introducción a SQL

SQL (Structured Query Language) es el lenguaje estándar para interactuar con bases de datos.

#### Ejemplo: Crear una tabla de productos.

```sql
CREATE TABLE productos (
    id INTEGER PRIMARY KEY,
    nombre TEXT NOT NULL,
    precio REAL
);
```

## 3. Integración de Python con bases de datos

Python puede conectarse a bases de datos usando librerías como `sqlite3` para manejar datos económicos.

#### Ejemplo: Crear una base de datos SQLite y agregar productos.

```python
import sqlite3

# Conectar a la base de datos (o crearla)
conn = sqlite3.connect('economia.db')
cursor = conn.cursor()

# Crear tabla
cursor.execute('''
CREATE TABLE IF NOT EXISTS productos (
    id INTEGER PRIMARY KEY,
    nombre TEXT NOT NULL,
    precio REAL
)
''')

# Insertar datos
productos = [
    (1, 'Producto A', 10.5),
    (2, 'Producto B', 20.0),
    (3, 'Producto C', 15.75)
]
cursor.executemany('INSERT OR REPLACE INTO productos VALUES (?, ?, ?)', productos)
conn.commit()

# Consultar datos
cursor.execute('SELECT * FROM productos')
for fila in cursor.fetchall():
    print(fila)

conn.close()
```

## 4. Mini análisis: ¿Por qué usar bases de datos?

Permiten almacenar grandes volúmenes de datos estructurados, facilitando consultas rápidas y seguras.
