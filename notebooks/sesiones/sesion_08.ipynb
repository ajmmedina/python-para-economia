# 1. Introducción a la visualización de datos

La visualización de datos es fundamental para entender y comunicar información económica de forma clara y atractiva.

#### Ejemplo: Distribución de ventas por día:

```python
import matplotlib.pyplot as plt

ventas = [1000, 1050, 1025, 1100]
dias = ["Lunes", "Martes", "Miércoles", "Jueves"]

plt.plot(dias, ventas)
plt.title("Ventas diarias")
plt.xlabel("Día")
plt.ylabel("Monto en S/")
plt.show()
```

## 2. Gráficos de barras para comparar datos

Los gráficos de barras son útiles para comparar cantidades en diferentes categorías.

#### Ejemplo: Distribución de ventas por producto.

```python
productos = ["Producto A", "Producto B", "Producto C"]
ventas = [1500, 2300, 1800]

plt.bar(productos, ventas, color='skyblue')
plt.title("Ventas por producto")
plt.xlabel("Producto")
plt.ylabel("Ventas en S/")
plt.show()
```

## 3. Gráficos de pastel para distribución porcentual

Los gráficos de pastel muestran la proporción de cada categoría respecto al total.

#### Ejemplo: Distribución del gasto mensual.

```python
categorias = ["Alquiler", "Comida", "Transporte", "Ocio"]
gastos = [1200, 800, 300, 400]

plt.pie(gastos, labels=categorias, autopct='%1.1f%%', startangle=90)
plt.title("Distribución del gasto mensual")
plt.show()
```

## 4. Mini análisis: ¿Por qué usar visualizaciones?
