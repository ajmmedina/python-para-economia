## 1. Repaso rápido de POO

Recordamos los conceptos básicos: clases, objetos, atributos, métodos, herencia y encapsulamiento.

#### Ejemplo:

```python
class Producto:
    def __init__(self, nombre, precio):
        self.nombre = nombre
        self.precio = precio

    def mostrar_info(self):
        print(f"Producto: {self.nombre}, Precio: S/ {self.precio}")

producto = Producto("Café", 15)
producto.mostrar_info()
```

## 2. Polimorfismo

El polimorfismo permite que diferentes clases tengan métodos con el mismo nombre, pero comportamientos distintos.

#### Ejemplo:

```python
class Producto:
    def descripcion(self):
        return "Producto genérico"

class ProductoPerecible(Producto):
    def descripcion(self):
        return "Producto perecible con fecha de vencimiento"

class ProductoNoPerecible(Producto):
    def descripcion(self):
        return "Producto no perecible"

productos = [Producto(), ProductoPerecible(), ProductoNoPerecible()]
for p in productos:
    print(p.descripcion())
```

## 3. Métodos especiales (mágicos)

Son métodos con nombres especiales que permiten definir comportamientos personalizados.

#### Ejemplo: Método `__str__` para mostrar información legible.

```python
class Cliente:
    def __init__(self, nombre, saldo):
        self.nombre = nombre
        self.saldo = saldo

    def __str__(self):
        return f"Cliente: {self.nombre}, Saldo: S/ {self.saldo}"

cliente = Cliente("Luis", 5000)
print(cliente)
```

## 4. Aplicación práctica: Sistema simple de cartera de inversiones

Creamos clases para representar activos financieros y una cartera que los contiene.

#### Ejemplo:

```python
class Activo:
    def __init__(self, nombre, valor):
        self.nombre = nombre
        self.valor = valor

    def valor_actual(self):
        return self.valor

class Accion(Activo):
    def __init__(self, nombre, valor, dividendos):
        super().__init__(nombre, valor)
        self.dividendos = dividendos

    def valor_actual(self):
        return self.valor + self.dividendos

class Cartera:
    def __init__(self):
        self.activos = []

    def agregar_activo(self, activo):
        self.activos.append(activo)

    def valor_total(self):
        return sum([a.valor_actual() for a in self.activos])

# Uso
cartera = Cartera()
cartera.agregar_activo(Accion("Empresa A", 1000, 50))
cartera.agregar_activo(Activo("Bonos", 2000))
print(f"Valor total de la cartera: S/ {cartera.valor_total()}")
```

## 5. Manejo de excepciones en POO

Cómo manejar errores y excepciones en métodos.

#### Ejemplo:

```python
class CuentaBancaria:
    def __init__(self, titular, saldo):
        self.titular = titular
        self.saldo = saldo

    def retirar(self, monto):
        if monto > self.saldo:
            raise ValueError("Saldo insuficiente")
        self.saldo -= monto

cuenta = CuentaBancaria("Ana", 1000)
try:
    cuenta.retirar(1500)
except ValueError as e:
    print(e)
```
