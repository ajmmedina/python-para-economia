## 1. ¿Qué son las series temporales?

Las series temporales son datos recogidos en intervalos regulares de tiempo, muy comunes en economía para analizar tendencias, ciclos y estacionalidades.

#### Ejemplo:

```python
import pandas as pd

fechas = pd.date_range(start='2023-01-01', periods=6, freq='M')
datos = [100, 120, 130, 125, 140, 150]
serie = pd.Series(datos, index=fechas)
print(serie)
```

## 2. Visualización de series temporales

Usamos `matplotlib` para graficar y observar patrones.

#### Ejemplo:

```python
import matplotlib.pyplot as plt

serie.plot()
plt.title('Ventas Mensuales')
plt.xlabel('Fecha')
plt.ylabel('Ventas')
plt.show()
```

## 3. Descomposición de series temporales

Separar la serie en tendencia, estacionalidad y ruido usando statsmodels.

#### Ejemplo:

```python
from statsmodels.tsa.seasonal import seasonal_decompose

resultado = seasonal_decompose(serie, model='additive')
resultado.plot()
plt.show()
```

## 4. Suavizado y medias móviles

Aplicar medias móviles para suavizar fluctuaciones.

#### Ejemplo:

```python
media_movil = serie.rolling(window=3).mean()
media_movil.plot()
plt.title('Media Móvil de Ventas')
plt.show()
```

## 5. Predicción básica con ARIMA

Introducción a modelos ARIMA para predicción.

#### Ejemplo:

```python
from statsmodels.tsa.arima.model import ARIMA

modelo = ARIMA(serie, order=(1,1,1))
modelo_fit = modelo.fit()
prediccion = modelo_fit.forecast(steps=3)
print(prediccion)
```
