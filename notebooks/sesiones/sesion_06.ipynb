## 1. Introducción a las listas

Las listas son colecciones ordenadas y mutables que permiten almacenar múltiples elementos, como números, textos o incluso otras listas.

#### Ejemplo:

```python
precios = [10.5, 12.0, 11.8, 13.2]
print(precios)
```

#### Ejemplo aplicado a economía: 

Lista con precios de un producto en diferentes meses.

```python
precios_mensuales = [1000, 1050, 1025, 1100, 1080]
print("Precios mensuales:", precios_mensuales)
```

## 2. Operaciones básicas con listas

- Acceder a elementos por índice.

- Modificar elementos.

- Agregar y eliminar elementos.

#### Ejemplo:

```python
precios_mensuales[2] = 1030  # Modifica el tercer mes
precios_mensuales.append(1120)  # Agrega un nuevo mes
print(precios_mensuales)
```

## 3. Introducción a los diccionarios

Los diccionarios almacenan pares clave-valor, ideales para representar datos con etiquetas, como nombres y valores asociados.

#### Ejemplo:

```python
producto = {"nombre": "Manzana", "precio": 3.5, "stock": 150}
print(producto)
```

#### Ejemplo aplicado a economía:

Diccionario con información de un producto.

```python
venta = {
    "producto": "Laptop", 
    "cantidad": 5, 
    "precio_unitario": 2500
}
print(f"Venta: {venta['cantidad']} unidades de {venta['producto']} a S/ {venta['precio_unitario']} cada una.")
```

## 4. Acceso y modificación en diccionarios

- Acceder a valores por clave.

- Modificar valores.

- Agregar y eliminar pares clave-valor.

#### Ejemplo:

```python
venta["cantidad"] = 7  # Modifica cantidad
venta["descuento"] = 0.1  # Agrega un nuevo campo
print(venta)
```

## 5. Mini análisis: ¿Por qué usar listas y diccionarios?

Estas estructuras permiten organizar datos de forma eficiente, facilitando su manipulación y análisis, especialmente en contextos económicos donde la información suele estar etiquetada y ordenada.
