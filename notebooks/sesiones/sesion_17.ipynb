## 1. Variables y parámetros financieros

Ejemplo:

```python
# Definir variables financieras
inversion_inicial = 10000
flujos = [3000, 4000, 5000, 6000]
tasa_descuento = 0.1
```

## 2. Funciones para cálculos financieros

Ejemplo:

```python
def vpn(tasa, flujos):
    return sum([f / (1 + tasa)**i for i, f in enumerate(flujos)])

valor_presente = vpn(tasa_descuento, flujos)
print(f"Valor Presente Neto: {valor_presente:.2f}")
```

## 3. Simulación de escenarios

Ejemplo:

```python
# Cambiar tasa de descuento y evaluar
for tasa in [0.08, 0.1, 0.12]:
    vpn_valor = vpn(tasa, flujos)
    print(f"Tasa: {tasa}, VPN: {vpn_valor:.2f}")
```
