## 1. ¿Qué es una función?

Una función es un bloque de código que realiza una tarea específica y puede ser reutilizado en diferentes partes del programa. Las funciones ayudan a organizar el código, hacerlo más legible y evitar la repetición.

#### Ejemplo:

```python
def saludar():
    print("¡Hola! Bienvenido a la clase de Python para Economía.")

saludar()

## 2. Definir funciones con parámetros

Las funciones pueden recibir datos de entrada llamados parámetros para trabajar con ellos dentro de la función.

#### Ejemplo:

```python
def saludar(nombre):
    print("Hola, " + nombre + "! Bienvenido a la clase de Python.")

saludar("Ana")
```

#### Ejemplo aplicado a economía: 

Función que calcula el ingreso total dado el precio y la cantidad vendida.

```python
def calcular_ingreso(precio, cantidad):
    ingreso = precio * cantidad
    return ingreso

resultado = calcular_ingreso(50, 200)
print(f"El ingreso total es S/ {resultado}")
```

## 3. Funciones que devuelven valores

Las funciones pueden devolver resultados usando la palabra clave `return`.

#### Ejemplo:

```python
def sumar(a, b):
    return a + b

resultado = sumar(3, 5)
print("La suma es", resultado)
```

#### Ejemplo aplicado a economía: 

Función que calcula el impuesto sobre una venta.

```python
def calcular_impuesto(venta, tasa_impuesto):
    impuesto = venta * tasa_impuesto
    print(f"El impuesto sobre la venta de S/ {venta} es S/ {impuesto}")

calcular_impuesto(1000, 0.18)
```

#### Ejemplo aplicado a economía: 

Función que calcula el promedio de ventas.

```python
def promedio_ventas(ventas):
    total = sum(ventas)
    cantidad = len(ventas)
    return total / cantidad

ventas = [100, 200, 150, 300, 250]
promedio = promedio_ventas(ventas)
print(f"El promedio de ventas es S/ {promedio}")
```

## 4. Mini análisis: ¿Por qué usar funciones?

Las funciones permiten dividir un programa en partes más pequeñas y manejables, facilitando la comprensión, el mantenimiento y la reutilización del código.
