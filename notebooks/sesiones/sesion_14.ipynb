## 1. Funciones de orden superior

Son funciones que reciben otras funciones como argumentos o devuelven funciones.

#### Ejemplo:

```python
def aplicar_funcion(func, lista):
    return [func(x) for x in lista]

ventas = [100, 200, 150]
resultado = aplicar_funcion(lambda x: x * 1.2, ventas)
print(resultado)  # Salida: [120.0, 240.0, 180.0]
```

## 2. Closures y decoradores

- **Closure:** Función que recuerda el entorno donde fue creada.

- **Decorador:** Función que modifica el comportamiento de otra función.

#### Ejemplo:

```python
def decorador_imprimir(func):
    def wrapper(*args, **kwargs):
        print("Ejecutando función...")
        resultado = func(*args, **kwargs)
        print(f"Resultado: {resultado}")
        return resultado
    return wrapper

@decorador_imprimir
def calcular_ingreso(precio, cantidad):
    return precio * cantidad

calcular_ingreso(50, 200)
```

## 3. Recursión

Función que se llama a sí misma para resolver problemas.

#### Ejemplo: Cálculo de factorial.

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

print(factorial(5))  # Salida: 120
```
