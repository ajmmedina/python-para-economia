## 1. Introducción a las estructuras condicionales

Las estructuras condicionales permiten que un programa tome decisiones basadas en condiciones que se evalúan como verdaderas o falsas. La estructura más común es el `if`, que ejecuta un bloque de código solo si la condición es verdadera.

#### Ejemplo básico:

```python
edad = 18
if edad >= 18:
    print("Eres mayor de edad.")
```

## 2. Uso de `else` y `elif`

Además de `if`, podemos usar `else` para ejecutar código cuando la condición es falsa, y `elif` para evaluar múltiples condiciones.

#### Ejemplo:

```python
edad = 16
if edad >= 18:
    print("Eres mayor de edad.")
elif edad >= 13:
    print("Eres un adolescente.")
else:
    print("Eres un niño.")
```

## 3. Operadores de comparación y lógicos

Para construir condiciones, usamos operadores como:

- `==` igual a

- `!=` diferente de

- `>` mayor que

- `<` menor que

- `>=` mayor o igual que

- `<=` menor o igual que

También operadores lógicos para combinar condiciones:

- `and` (y)

- `or` (o)

- `not` (no)

#### Ejemplo:

```python
nota = 85
if nota >= 70 and nota < 90:
    print("Aprobaste con buena nota.")
```
