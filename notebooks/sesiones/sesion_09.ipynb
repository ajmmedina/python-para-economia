## 1. Introducción a la visualización avanzada

La visualización avanzada permite explorar datos económicos complejos y descubrir patrones que no son evidentes a simple vista.

#### Ejemplo: Graficar la relación entre variables económicas usando `seaborn`.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Datos de ejemplo: precios y demanda de un producto
precios = [10, 12, 15, 20, 25, 30]
demanda = [100, 90, 80, 60, 50, 40]

sns.scatterplot(x=precios, y=demanda)
plt.title("Relación entre precio y demanda")
plt.xlabel("Precio (S/)")
plt.ylabel("Demanda (unidades)")
plt.show()
```

## 2. Análisis con `numpy` y `pandas`

Estas librerías permiten manipular y analizar grandes conjuntos de datos económicos de forma eficiente.

#### Ejemplo: Calcular estadísticas básicas de ventas con `pandas`.

```python
import pandas as pd

# Crear un DataFrame con datos de ventas
ventas = pd.DataFrame({
    'Producto': ['A', 'B', 'C', 'D'],
    'Ventas': [1500, 2300, 1800, 2000]
})

# Calcular estadísticas descriptivas
print(ventas['Ventas'].describe())
```

## 3. Gráficos avanzados con seaborn

`Seaborn` facilita la creación de gráficos estadísticos para análisis económico.

#### Ejemplo: Boxplot para analizar la distribución de ventas por producto.

```python
sns.boxplot(x='Producto', y='Ventas', data=ventas)
plt.title("Distribución de ventas por producto")
plt.show()
```
