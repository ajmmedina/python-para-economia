## 1. Introducción a los bucles

Los bucles permiten ejecutar un bloque de código varias veces de forma automática, lo que es útil para tareas repetitivas sin tener que escribir el mismo código muchas veces.

En Python, los bucles más comunes son `for` y `while`.

#### Ejemplo: 

Un bucle `for` que imprime números del 1 al 5:

```python
for i in range(1, 6):
    print(i)
```

#### Ejemplo aplicado a economía: 

Imprimir los precios de un producto durante 5 días consecutivos.

```python
precios = [10.5, 10.7, 10.6, 10.8, 11.0]
for precio in precios:
    print(f"Precio del producto: S/ {precio}")
```

## 2. Uso del bucle while

El bucle `while` ejecuta un bloque de código mientras una condición sea verdadera.

#### Ejemplo:

```python
contador = 1
while contador <= 5:
    print(contador)
    contador += 1
```

#### Ejemplo aplicado a economía: 

Solicitar al usuario que ingrese ventas diarias hasta que ingrese un valor negativo para terminar.

```python
ventas = []
venta = 0
while venta >= 0:
    venta = float(input("Ingrese la venta diaria (número negativo para terminar): "))
    if venta >= 0:
        ventas.append(venta)
print(f"Ventas registradas: {ventas}")
```

## 3. Diferencias entre `for` y `while`

- `for` se usa cuando sabemos cuántas veces queremos repetir el bloque (iterar sobre una secuencia).

- `while` se usa cuando queremos repetir mientras una condición se cumpla, sin un número fijo de repeticiones.

## 4. Mini análisis: ¿Por qué usar bucles?

Los bucles hacen que el código sea más eficiente y fácil de mantener, evitando la repetición innecesaria y permitiendo automatizar tareas repetitivas.
