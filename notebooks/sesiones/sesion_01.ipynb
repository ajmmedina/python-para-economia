## 1. Introducción a la indentación

La indentación en Python es fundamental porque indica al intérprete qué bloques de código pertenecen juntos. A diferencia de otros lenguajes que usan llaves o palabras clave para delimitar bloques, Python utiliza espacios o tabulaciones para definir la estructura del código. Si la indentación no es correcta, el programa no funcionará y mostrará errores.

#### Ejemplo: 

Código con indentación incorrecta que genera un error.

```python
if 5 > 2:
print("Cinco es mayor que dos")  # Error de indentación
```

#### Ejemplo: 

Código corregido con indentación adecuada.

```python
if 5 > 2:
    print("Cinco es mayor que dos")
```

## 2. Uso de `print()` para mostrar mensajes

La función `print()` se usa para mostrar información en la pantalla. Es una de las funciones más básicas y útiles para comunicar resultados o mensajes al usuario.

#### Ejemplos:

```python
print("Hola, mundo!")
print("Aprendiendo Python para economía")
```

## 3. Uso de `input()` para pedir datos al usuario

La función `input()` permite al programa recibir datos que el usuario escribe. Esto hace que el programa sea interactivo.

#### Ejemplo: 

Pide el nombre del usuario y mostrar un saludo personalizado.

```python
nombre = input("¿Cuál es tu nombre? ")
print("Hola, " + nombre + "! Bienvenido a la clase de Python para economía.")
```

## 4. Mini análisis: ¿Qué pasa si no respetamos la indentación?

Cuando no respetamos la indentación, Python muestra errores de sintaxis que impiden que el programa se ejecute. Estos errores ayudan a detectar problemas en la estructura del código. Mantener una sintaxis clara y ordenada evita estos errores y facilita la lectura y mantenimiento del código.

## 7. Reflexión: ¿Por qué la sintaxis clara es importante en programación?

Una sintaxis clara y correcta permite que el código sea entendido no solo por la computadora, sino también por otros programadores y por uno mismo en el futuro. Facilita la detección de errores, la colaboración y la ampliación del código. En Python, la indentación es parte de esa claridad, haciendo que el código sea legible y organizado.
