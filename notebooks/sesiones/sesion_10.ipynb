# 1. ¿Qué es la programación orientada a objetos?

La programación orientada a objetos (POO) es un paradigma que organiza el código en "objetos" que representan entidades del mundo real, con atributos (propiedades) y métodos (funciones).

#### Ejemplo:

```python
class Producto:
    def __init__(self, nombre, precio):
        self.nombre = nombre
        self.precio = precio

    def mostrar_info(self):
        print(f"Producto: {self.nombre}, Precio: S/ {self.precio}")

producto1 = Producto("Manzana", 3.5)
producto1.mostrar_info()
```

## 2. Clases y objetos

Una clase es un molde para crear objetos. Un objeto es una instancia de una clase con valores específicos.

#### Ejemplo: Clase para representar un cliente y calcular su gasto total.

```python
class Cliente:
    def __init__(self, nombre, compras):
        self.nombre = nombre
        self.compras = compras  # lista de montos

    def gasto_total(self):
        return sum(self.compras)

cliente1 = Cliente("Ana", [100, 200, 150])
print(f"Gasto total de {cliente1.nombre}: S/ {cliente1.gasto_total()}")
```

## 3. Métodos y atributos

Los atributos son variables que describen al objeto. Los métodos son funciones que definen comportamientos.

#### Ejemplo: Clase para aplicar descuento a un producto.

```python
class Producto:
    def __init__(self, nombre, precio):
        self.nombre = nombre
        self.precio = precio

    def aplicar_descuento(self, porcentaje):
        self.precio *= (1 - porcentaje/100)

producto = Producto("Laptop", 2500)
producto.aplicar_descuento(10)
print(f"Precio con descuento: S/ {producto.precio}")
```

## 4. Herencia

La herencia permite crear nuevas clases basadas en clases existentes, reutilizando código.

#### Ejemplo: Clase para productos perecibles que hereda de Producto.

```python
class ProductoPerecible(Producto):
    def __init__(self, nombre, precio, fecha_vencimiento):
        super().__init__(nombre, precio)
        self.fecha_vencimiento = fecha_vencimiento

    def mostrar_info(self):
        print(f"Producto: {self.nombre}, Precio: S/ {self.precio}, Vence: {self.fecha_vencimiento}")

producto_p = ProductoPerecible("Leche", 5.0, "2024-06-30")
producto_p.mostrar_info()
```

## 5. Encapsulamiento

El encapsulamiento protege los datos internos del objeto, permitiendo acceso controlado.

#### Ejemplo:

```python
class CuentaBancaria:
    def __init__(self, titular, saldo):
        self.titular = titular
        self.__saldo = saldo  # atributo privado

    def depositar(self, monto):
        if monto > 0:
            self.__saldo += monto

    def retirar(self, monto):
        if 0 < monto <= self.__saldo:
            self.__saldo -= monto

    def mostrar_saldo(self):
        print(f"Saldo de {self.titular}: S/ {self.__saldo}")

cuenta = CuentaBancaria("Carlos", 1000)
cuenta.depositar(500)
cuenta.retirar(200)
cuenta.mostrar_saldo()
```
