# 1. ¿Qué es la programación funcional?

La programación funcional es un paradigma que trata la computación como la evaluación de funciones matemáticas y evita cambiar estados o datos.

#### Ejemplo:

```python
# Función lambda para sumar dos números
suma = lambda x, y: x + y
print(suma(5, 3))  # Salida: 8
```

## 2. Funciones lambda

Son funciones anónimas, útiles para operaciones simples y rápidas.

#### Ejemplo:

```python
# Calcular ingresos usando lambda
calcular_ingreso = lambda precio, cantidad: precio * cantidad
print(calcular_ingreso(50, 200))  # Salida: 10000
```

## 3. Funciones `map`, `filter` y `reduce`

- `map`: Aplica una función a todos los elementos de una lista.

- `filter`: Filtra elementos de una lista según una condición.

- `reduce`: Aplica una función acumulativa a los elementos de una lista.

#### Ejemplo:

```python
from functools import reduce

ventas = [100, 200, 150, 300, 250]

# Map: aumentar ventas en 10%
ventas_aumentadas = list(map(lambda x: x * 1.1, ventas))

# Filter: ventas mayores a 200
ventas_altas = list(filter(lambda x: x > 200, ventas))

# Reduce: suma total de ventas
ventas_totales = reduce(lambda x, y: x + y, ventas)

print(ventas_aumentadas)
print(ventas_altas)
print(ventas_totales)
```
