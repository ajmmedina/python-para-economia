## 1. Variables numéricas, cadenas y booleanos

En Python, una variable es un espacio en la memoria donde almacenamos datos. Los tipos más comunes son:

- Números enteros (`int`), por ejemplo: 10, -3, 0

- Números decimales (`float`), por ejemplo: 3.14, -0.5

- Cadenas de texto (`str`), por ejemplo: "Hola", 'Python'

- Booleanos (`bool`), que pueden ser `True` o `False`

#### Ejemplos:

```python
entero = 10
decimal = 3.14
texto = "Economía"
booleano = True
```

## 2. Operaciones básicas con números

Podemos realizar operaciones matemáticas básicas con variables numéricas:

```python
suma = 5 + 3
resta = 10 - 2
multiplicacion = 4 * 2
division = 8 / 2
```

## 3. Concatenación de cadenas

La concatenación es unir cadenas de texto usando el operador `+`:

```python
saludo = "Hola, " + "mundo!"
print(saludo) 
```

## 4. Introducción a listas, tuplas y diccionarios

Estas son estructuras para organizar datos:

- **Lista:** colección ordenada y mutable. Ejemplo:

```python
lista = [1, 2, 3, 4]
```

- **Tupla:** colección ordenada e inmutable. Ejemplo:

```python
tupla = ("a", "b", "c")
```

- **Diccionario:** colección de pares clave-valor. Ejemplo:

```python
diccionario = {"país": "Perú", "PIB": 2000, "inflación": 3.5}
```

## 5. Mini análisis: ¿Qué diferencia hay entre una lista y una tupla?

Las listas pueden modificarse (agregar, eliminar elementos), mientras que las tuplas no pueden cambiar una vez creadas. Esto es útil para datos que no deben alterarse.
