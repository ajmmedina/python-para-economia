# 1. Introducción al manejo de archivos

El manejo de archivos permite guardar y leer datos desde archivos externos, facilitando el almacenamiento y procesamiento de grandes volúmenes de información.

#### Ejemplo:

Escribir y leer un archivo de texto:

```python
# Escribir en un archivo
with open('ventas.txt', 'w') as archivo:
    archivo.write('''1000
1050
1025
1100''')

# Leer el archivo
with open('ventas.txt', 'r') as archivo:
    datos = archivo.readlines()
    print(datos)
```

#### Ejemplo aplicado a economía:

Guardar y leer ventas diarias desde un archivo.

```python
# Guardar ventas diarias en un archivo
ventas_diarias = [1000, 1050, 1025, 1100]
with open('ventas_diarias.txt', 'w') as archivo:
    for venta in ventas_diarias:
        archivo.write(str(venta) + "\n")

# Leer ventas desde el archivo
with open('ventas_diarias.txt', 'r') as archivo:
    ventas_leidas = archivo.readlines()
    ventas_leidas = [float(venta.strip()) for venta in ventas_leidas]
    print(f"Ventas leídas: {ventas_leidas}")
```

#### Ejemplo aplicado a la economía:

Crear un archivo de ventas diarias con dos columnas.

```python
with open('ventas.csv', 'w') as archivo: 
    archivo.write('fecha,monto ') 
    archivo.write('2024-01-01,1000 ') 
    archivo.write('2024-01-02,1050 ') 
    archivo.write('2024-01-03,1025 ') 
    archivo.write('2024-01-04,1100 ')
```

## 2. Uso de librerías externas

Python permite usar librerías externas para ampliar sus funcionalidades, como pandas para análisis de datos.

#### Ejemplo:

Librería `pandas` para leer un archivo CSV:

```python
import pandas as pd

ventas = pd.read_csv('ventas.csv')
print(ventas.head())
```

#### Ejemplo aplicado a economía: 

Analizar datos de ventas con pandas.

```python
import pandas as pd

# Supongamos que tenemos un archivo CSV con datos de ventas
ventas = pd.read_csv('ventas.csv')

# Mostrar las primeras filas
print(ventas.head())

# Calcular el total de ventas
total_ventas = ventas['monto'].sum()
print(f"Total de ventas: S/ {total_ventas}")

# Calcular promedio de ventas
promedio_ventas = ventas['monto'].mean()
print(f"Promedio de ventas: S/ {promedio_ventas}")
```
