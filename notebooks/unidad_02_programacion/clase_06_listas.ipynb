# Lección 6: Listas

Una lista permite guardar varios valores.

Ejemplo:

- inflacion = [2.1, 2.5, 3.0, 3.2]

## Escribe el código

>inflacion = [2.1, 2.5, 3.0, 3.2]

>print(inflacion)

## Acceder a datos

Podemos acceder a un valor usando su posición.

Ejemplo:

>print(inflacion[0])
