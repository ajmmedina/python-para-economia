# Lección 5: Bucles

Un bucle permite repetir código varias veces.

En Python usamos `for`.

## Ejemplo

Escribe el siguiente código:

>for i in range(5):

>&emsp;&emsp;print("Análisis económico")

## Explicación

- `range(5)` es una secuencia de 5 números enteros que comienza en cero y termina en cuatro (es decir $0,1,2,3,4$)
- `for` recorre toda la secuencia `range(5)` y por cada valor imprime el mensaje *Análisis económico*
