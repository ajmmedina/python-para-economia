# Lección 4: Condicionales

En economía muchas decisiones dependen de condiciones.

Por ejemplo:

- Si el ingreso es mayor que el costo → hay ganancia.

Python permite evaluar condiciones usando `if`.

## Ejemplo

Escribe el siguiente código (no te olvides del espaciado antes del `print`):

>ingreso = 100

>costo = 80

>if ingreso > costo:

>&emsp;&emsp;print("Hay ganancia")

## Explicación

La condición:

- ingreso > costo

significa:

- ¿el ingreso es mayor que el costo?

## Modifica el programa

Cambia los valores:

>ingreso = 50

>costo = 80

¿Qué ocurre?

## Reto económico

Crea un programa que diga:

>"El negocio tiene ganancias"

>si ingreso > costo
