# Lección 7: Funciones

Una función es un bloque de código reutilizable.

Ejemplo:

>def ingreso(precio, cantidad):

>&emsp;&emsp;resultado = precio * cantidad

>&emsp;&emsp;return resultado

### La función en acción

>print(ingreso(10,5))

### Explicación

La función calcula:

$$Ingreso = precio × cantidad$$

- Si el precio es 10 y se venden 5 unidades, el ingreso será $50$.

## Crear función beneficio

La función beneficio se calcula:

$$beneficio = ingreso - costo$$

### Escriba el siguiente código:

>def beneficio(precio, cantidad, costo):

>&emsp;&emsp;ingreso_total = precio * cantidad

>&emsp;&emsp;resultado = ingreso_total - costo

>&emsp;&emsp;return resultado

### Probar la función:

>print(beneficio(20,10,150))
