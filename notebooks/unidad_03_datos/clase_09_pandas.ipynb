# Lección 10: Tablas de datos económicos

_**🎯 Objetivo:** Trabajar con tablas de datos usando Pandas._

Los economistas organizan los datos en tablas:

Año | Inflación | PIB

## Paso 1: Importar la librería `Pandas`

`Pandas` es una biblioteca de código abierto para Python, fundamental en la ciencia de datos, especializada en la manipulación, limpieza y análisis de datos estructurados. Utiliza estructuras de alto rendimiento como DataFrame (tablas, similar a Excel) y Series, permitiendo trabajar eficientemente con grandes volúmenes de datos numéricos y temporales.

### Escribe el siguiente código:

>import pandas as pd

>import matplotlib.pyplot as plt

#### Recuerda que:

La librería `Pandas` no viene instalada por defecto en Python, hay que instalarla con el comando `pip`.

>%pip install pandas

## Paso 2: Crear vectores

Los vectores se pueden crear desde Numpy o desde listas de Python

### Escribe el siguiente código:

>años = [2018,2019,2020,2021,2022]

>inflacion = [2.2,2.1,2.0,6.4,8.5]

>pib = [4.0,2.2,-11.0,13.3,2.7]

## Paso 3: Crear un Dataframe

Un Dataframe es muy similar a una tabla de Excel.

### Escribe el siguiente código:

>datos = pd.DataFrame({

>&emsp;&emsp;"Año":años,

>&emsp;&emsp;"Inflacion":inflacion,

>&emsp;&emsp;"PIB":pib

>})

## Paso 4: Mostrar datos

- Los datos se pueden mostrar con la intrucción `print`.
- En un cuaderno `notebook` los datos se muestran simplemente escribiendo el nombre de la variable al final de la celda de trabajo.

## Paso 5: Graficar los datos

Para graficar usaremos la librería `matplotlib`.

### Escribe el siguiente código:

>plt.plot(datos["Año"],datos["Inflacion"])

>plt.xlabel("Año")

>plt.ylabel("Inflación (%)")

>plt.title("Inflación en Perú")

>plt.show()
