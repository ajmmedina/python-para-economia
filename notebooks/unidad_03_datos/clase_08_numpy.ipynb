## Paso 1: Importar la librería `Numpy`

`Numpy` es una biblioteca fundamental para la computación científica en Python, diseñada para manipular grandes arreglos y matrices multidimensionales de forma eficiente. Ofrece herramientas matemáticas de alto nivel, siendo hasta 50 veces más rápida que las listas tradicionales de Python, lo que la hace indispensable en análisis de datos, IA y machine learning.

### Escribe el siguiente código:

>import numpy as np

#### Recuerda que:

La librería `Numpy` no viene instalada por defecto en Python, hay que instalarla con el comando `pip`.

>%pip install numpy

## Paso 2: Crear un vector de inflación

Un vector es una colección de números. La instrucción `np.array` crea un vector en Numpy.

### Escribe el siguiente código:

>inflacion = np.array([2.1, 2.5, 3.0, 4.2, 3.8])

## Paso 3: Mostrar los datos

Para mostrar los datos usamos `print`.

### Escribe el siguiente código:

>print(inflacion)

## Paso 4: Promedio de inflación

Numpy permite calcular el promedio de un conjunto de datos usando la instrucción `np.mean`.

### Escribe el siguiente código:

>inflacion_promedio = np.mean(inflacion)

>print(inflacion_promedio)

## Paso 5: Inflación máxima

Numpy calcula el máximo valor de un conjunto de datos usando la instrucción `np.max`.

### Escribe el siguiente código:

>inflacion_maxima = np.max(inflacion)

>print(inflacion_maxima)

#### A tener en cuenta

- En Numpy se puede modificar los elementos de un vector de la misma forma como se hace en python.

>vector = np.array([1, 8, 12])

>vector[0] = 25

>vector tendrá los elementos [25, 8, 12]

- En Numpy no se puede aumentar elementos a un vector, pero se puede crear un nuevo vector con el vector anterior y los nuevos elementos.

>vector = np.array([2, 3, 5])

>vector = np.append(vector, [7, 8])

>vector tendrá los elementos [2, 3, 5, 7, 8]
