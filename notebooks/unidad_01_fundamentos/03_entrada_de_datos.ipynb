# Lección 3: Entrada de datos

Podemos pedir información al usuario usando `input()`.

Ejemplo:

>nombre = input("Escribe tu nombre: ")

>print(nombre)

## Entrada de números

Cuando usamos `input()` el dato es texto.

Para usar números debemos convertirlos.

Ejemplo:

>precio = float(input("Precio: "))

>cantidad = int(input("Cantidad: "))

>ingreso = precio * cantidad

>print(ingreso)
