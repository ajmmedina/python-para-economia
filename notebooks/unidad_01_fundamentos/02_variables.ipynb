# Lección 2: Variables

Una variable es un espacio donde guardamos información.

Ejemplo:

>precio = 10

## Escribe el código

>precio = 10

>cantidad = 5

>print(precio)

>print(cantidad)

## Cálculos con variables

Python puede hacer operaciones matemáticas.

Ejemplo:

>precio = 10

>cantidad = 5

>ingreso = precio * cantidad

>print(ingreso)

Escribe ese código.

## Interpretación económica

Si el precio es 10 y se venden 5 unidades,
el ingreso total es:

$$ingreso = precio × cantidad$$

## Reto

Una empresa vende un producto a 25 dólares.

Vendió 12 unidades.

Crea un programa que calcule el ingreso total.
