# Proyecto: Datos del Banco Central de Reserva del Perú

En este proyecto descargaremos datos económicos reales.

Usaremos la API del BCRP para obtener series de datos.

#### API pública del Banco Central de Reserva del Perú

El [Banco Central de Reserva del Perú](https://estadisticas.bcrp.gob.pe/estadisticas/series/) tiene una API pública de datos económicos.

Los economistas pueden descargar datos como:

- Inflación
- Tipo de cambio
- Tasas de interés
- PBI
- Reservas internacionales

## Paso 1: Importar librerías

Las librerías necesarias para este proyecto son:

- `pandas`: para manipular y analizar datos.
- `matplotlib.pyplot`: para graficar datos.
- `requests`: para obtener datos desde internet.



### Escribe el siguiente código:

>import pandas as pd

>import requests

>import matplotlib.pyplot as plt

## Paso 2: Definir la serie económica

Por ejemplo analizaremos ***inflación***

### Escribe el siguiente código:

>url = "https://estadisticas.bcrp.gob.pe/estadisticas/series/api/PN01273PM/json"

## Paso 3: Descargar datos

Usamos la librería `requests` para la descarga de datos.

### Escribe el siguiente código:

>response = requests.get(url)

>data = response.json()

## Paso 4: Convertir datos en tabla

Usamos la librería `pandas` para analizar ***periods***.

Convertimos los datos en una tabla `Dataframe`.

### Escribe el siguiente código:

>datos = data["periods"]

>df = pd.DataFrame(datos)

## Paso 5: Preparar los datos

Realizamos una limpieza a los datos, convertimos a tipo `float` la columna ***value***

### Escribe el siguiente código:

>df["value"] = df["value"].astype(float)

## Paso 6: Mostrar datos

Debido a la gran cantidad de datos que puede tener un `Dataframe`, mostraremos los 5 primeros registros.

### Escribe el siguiente código:

>print(df.head())

## Paso 7: Crear gráfico

Creamos un gráfico con las dos únicas columnas ***name*** y ***value***

### Escribe el siguiente código:

>plt.plot(df["name"], df["value"])

>plt.title("Inflación - Perú")

>plt.xlabel("Periodo")

>plt.ylabel("Inflación")

>plt.show()
