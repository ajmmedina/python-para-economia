# Proyecto 1: Calculadora de interés compuesto

El interés compuesto se calcula con la fórmula:

$$Monto = Capital \cdot (1 + tasa)^{años}$$

Este programa calculará cuánto dinero tendrás después de varios años.

## Paso 1

El programa debe pedir al usuario:

- capital inicial
- tasa de interés
- número de años

## Paso 2

Convierte la tasa a decimal.

Ejemplo:

- $5\% = 0.05$

## Paso 3

Calcula el monto final usando la fórmula:

$$Monto = capital \cdot (1 + tasa)^{años}$$

## Paso 4

Muestra el resultado.

## Reto
Calcula cuánto dinero tendrás después de:

- 10 años
- 20 años
- 30 años

>**Recomendación:** Los 4 pasos anteriores copialos en la siguiente celda y en esa celda resuelve el reto
