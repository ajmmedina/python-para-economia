# **Proyecto:** Manejo de bases de datos con SQLite y Python

**Propósito**

Aprenderás a crear una base de datos SQLite, insertar datos económicos y realizar consultas básicas usando Python.

## Pasos

1. Crear una base de datos SQLite llamada `economia.db`.

2. Crear una tabla `Indicadores` con columnas para nombre, valor y fecha.

3. Insertar varios registros con indicadores económicos (inflación, PIB, desempleo).

4. Consultar y mostrar todos los registros.

5. Actualizar un valor y eliminar un registro.

6. Cerrar la conexión a la base de datos.

## Reflexión

- Este proyecto te introduce al manejo de bases de datos, fundamental para almacenar y gestionar grandes volúmenes de datos económicos.

## Retos adicionales

1. Crear tablas adicionales para diferentes tipos de indicadores.

2. Realizar consultas con filtros y ordenamientos.

3. Implementar transacciones para operaciones seguras.

4. Exportar datos consultados a archivos CSV.

5. Integrar consultas con visualizaciones.

## ¿Qué aprendimos con este proyecto?

- Cómo crear y conectar con bases de datos SQLite desde Python.

- Creación de tablas y manipulación de datos (insertar, consultar, actualizar, eliminar).

- La importancia de bases de datos para el manejo estructurado de información.

- Buenas prácticas para cerrar conexiones y manejar errores.
