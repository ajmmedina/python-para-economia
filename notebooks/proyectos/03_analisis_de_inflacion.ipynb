# Proyecto 3: Análisis de inflación

Supongamos que tenemos la inflación de varios años.

Queremos calcular la inflación promedio.

**Observación:** La fórmula para calcular el promedio es

$$\bar{x}=\frac{\sum_{i=1}^n x_i}{n}$$

## Paso 1.

Crea una lista con inflaciones de varios años.

## Paso 2.

Calcule el promedio.

- Para sumar valores de una lista use `sum()`.
- Para saber la cantidad de elementos de una lista use `len()`.

## Paso 3.

Muestre el resultado con un contexto adecuado.

## Reto

Si la inflación promedio es mayor que 3%

- mostrar: ***Inflación alta***
- de lo contrario mostrar: ***Inflación controlada***
