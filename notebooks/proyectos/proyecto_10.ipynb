# **Proyecto:** Análisis estadístico básico con `pandas` y `scipy`

**Propósito**

Aprenderás a realizar análisis estadísticos básicos y pruebas de hipótesis en datos económicos.

## Pasos

1. Cargar datos económicos con `pandas`.

2. Calcular estadísticas descriptivas.

3. Realizar pruebas t y chi-cuadrado con `scipy`.

4. Interpretar resultados estadísticos.

5. Documentar conclusiones.

## Reflexión

- El análisis estadístico es fundamental para validar hipótesis y tomar decisiones en economía.

## Retos adicionales

1. Aplicar pruebas no paramétricas.

2. Comparar grupos con ANOVA.

3. Visualizar resultados estadísticos.

4. Automatizar reportes estadísticos.

5. Integrar análisis con modelos predictivos.

## ¿Qué aprendimos con este proyecto?

- Estadísticas descriptivas con `pandas`.

- Pruebas de hipótesis con `scipy`.

- Interpretación de resultados estadísticos.
