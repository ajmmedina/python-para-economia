# **Proyecto 5:** Análisis simple de gasto mensual

**Propósito**

Este proyecto permite practicar listas, funciones y condicionales para analizar y clasificar gastos mensuales de un hogar.

## Paso 1: Definir listas con datos de gasto

```python
gasto_mensual = [1500, 1600, 1550, 1580, 1620, 1650]
meses = ["Ene", "Feb", "Mar", "Abr", "May", "Jun"]
```

## Paso 2: Calcular promedio de gasto

```python
def promedio(lista):
    return sum(lista) / len(lista)

promedio_gasto = promedio(gasto_mensual)
print(f"Gasto promedio mensual: S/{promedio_gasto:.2f}")
```

## Paso 3: Clasificar gasto

```python
if promedio_gasto > 1600:
    categoria = "Alto"
elif promedio_gasto >= 1550:
    categoria = "Moderado"
else:
    categoria = "Bajo"

print(f"Categoría de gasto: {categoria}")
```

## Paso 4: Guardar resultados en archivo

```python
with open("resultado_gasto.txt", "w") as archivo:
    archivo.write(f"Gasto promedio mensual: S/{promedio_gasto:.2f}")
    archivo.write(f"Categoría de gasto: {categoria}")
```

## Paso 5: Leer y mostrar archivo

```python
with open("resultado_gasto.txt", "r") as archivo:
    contenido = archivo.read()
    print("Contenido del archivo resultado_gasto.txt:")
    print(contenido)
```
