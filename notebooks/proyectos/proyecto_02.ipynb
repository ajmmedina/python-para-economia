# **Proyecto 2:** Cálculo de inflación promedio y clasificación simple

**Propósito**

Este proyecto refuerza el manejo de listas, funciones y condicionales para calcular la inflación promedio anual y clasificarla según niveles económicos.

## Paso 1: Definir listas con datos de inflación mensual

```python
inflacion_mensual = [0.2, 0.3, 0.25, 0.22, 0.28, 0.3, 0.27, 0.26, 0.29, 0.31, 0.33, 0.35]
meses = ["Ene", "Feb", "Mar", "Abr", "May", "Jun", "Jul", "Ago", "Sep", "Oct", "Nov", "Dic"]
```

## Paso 2: Crear función para calcular promedio

```python
def promedio(lista):
    return sum(lista) / len(lista)

promedio_inflacion = promedio(inflacion_mensual)
print(f"Inflación promedio anual: {promedio_inflacion:.2f}%")
```

## Paso 3: Clasificar inflación

```python
if promedio_inflacion > 0.3:
    categoria = "Alta"
elif promedio_inflacion >= 0.2:
    categoria = "Moderada"
else:
    categoria = "Baja"

print(f"Categoría de inflación: {categoria}")
```

## Paso 4: Guardar resultados en archivo

```python
with open("resultado_inflacion.txt", "w") as archivo:
    archivo.write(f"Inflación promedio anual: {promedio_inflacion:.2f}%")
    archivo.write(f"Categoría de inflación: {categoria}")
```

## Paso 5: Leer y mostrar archivo

```python
with open("resultado_inflacion.txt", "r") as archivo:
    contenido = archivo.read()
    print("Contenido del archivo resultado_inflacion.txt:")
    print(contenido)
```
