## **Proyecto:** Modelado de datos económicos con clases y herencia

**Propósito**

Diseñarás clases base y derivadas para representar diferentes tipos de datos económicos, aplicando herencia y polimorfismo.

## Pasos

1. Crear clase base `IndicadorEconomico` con atributos comunes.

2. Crear clases derivadas como `Inflacion`, `PIB`, `Desempleo` con atributos y métodos específicos.

3. Implementar métodos para calcular indicadores derivados.

4. Instanciar objetos y mostrar información.

5. Añadir métodos para comparar indicadores.

## Reflexión

- Este proyecto fortalece la capacidad de modelar datos complejos y relacionados en economía usando POO avanzada.

## Retos adicionales

1. Añadir métodos para validación de datos.

2. Implementar interfaces para exportar datos.

3. Crear clases para análisis estadístico.

4. Integrar con bases de datos.

5. Simular escenarios económicos con objetos.

## ¿Qué aprendimos con este proyecto?

- Diseño de clases base y derivadas.

- Uso de herencia y polimorfismo.

- Modelado de datos económicos con estructuras jerárquicas.

- Ventajas de POO para organizar código y datos.
