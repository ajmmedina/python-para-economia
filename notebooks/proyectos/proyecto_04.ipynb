# **Proyecto 4:** Cálculo de ahorro promedio mensual

**Propósito**

Este proyecto refuerza el uso de listas, funciones y condicionales para calcular y clasificar el ahorro mensual de una familia.

## Paso 1: Definir listas con datos de ahorro

```python
ahorro_mensual = [200, 250, 180, 220, 210, 230]
meses = ["Ene", "Feb", "Mar", "Abr", "May", "Jun"]
```

## Paso 2: Calcular promedio de ahorro

```python
def promedio(lista):
    return sum(lista) / len(lista)

promedio_ahorro = promedio(ahorro_mensual)
print(f"Ahorro promedio mensual: S/{promedio_ahorro:.2f}")
```

## Paso 3: Clasificar ahorro

```python
if promedio_ahorro > 220:
    categoria = "Alto"
elif promedio_ahorro >= 190:
    categoria = "Moderado"
else:
    categoria = "Bajo"

print(f"Categoría de ahorro: {categoria}")
```

## Paso 4: Guardar resultados en archivo

```python
with open("resultado_ahorro.txt", "w") as archivo:
    archivo.write(f"Ahorro promedio mensual: S/{promedio_ahorro:.2f}")
    archivo.write(f"Categoría de ahorro: {categoria}")
```

## Paso 5: Leer y mostrar archivo

```python
with open("resultado_ahorro.txt", "r") as archivo:
    contenido = archivo.read()
    print("Contenido del archivo resultado_ahorro.txt:")
    print(contenido)
```
