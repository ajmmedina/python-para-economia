# Proyecto 2: Simulador de crecimiento del PIB

El PIB de un país puede crecer cada año.

$$PIB_{futuro} = PIB_{actual} \cdot (1 + tasa_{crecimiento})$$

## Paso 1. 
El programa debe pedir:

- PIB actual
- tasa de crecimiento anual
- número de años

## Paso 2.
Convierte la tasa a decimal.

## Paso 3.
Usa un bucle para calcular el PIB cada año.

## Paso 4.
Haz que el programa también imprima el número de año.

*Resultado esperado*
Año 1: PIB ...
Año 2: PIB ...

## Reto
Calcula el PBI hasta:

- 10 años
- 15 años
- 20 años

>**Recomendación:** Los 4 pasos anteriores copialos en la siguiente celda y en esa celda resuelve el reto
