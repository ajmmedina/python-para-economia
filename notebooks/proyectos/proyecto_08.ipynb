# **Proyecto:** Manipulación avanzada de datos con `pandas`

**Propósito**

Aprenderás a filtrar, agrupar y transformar datos económicos para análisis más profundos.

## Pasos

1. Cargar un dataset económico con múltiples variables.

2. Filtrar datos según condiciones (fechas, valores).

3. Agrupar datos por categorías y calcular agregados.

4. Crear columnas nuevas con transformaciones.

5. Exportar los datos transformados.

## Reflexión

- Manipular datos es esencial para preparar información relevante en economía.

## Retos adicionales

1. Combinar datasets con `merge` o `join`.

2. Aplicar funciones personalizadas con `.apply()`.

3. Trabajar con datos temporales y fechas.

4. Detectar y manejar valores atípicos.

5. Automatizar procesos de limpieza y transformación.

## ¿Qué aprendimos con este proyecto?

- Filtrado y selección avanzada.

- Agrupamiento y agregación.

- Transformación y creación de variables.
