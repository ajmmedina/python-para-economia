**Propósito**

Este proyecto tiene como objetivo que el estudiante aplique conceptos fundamentales de Python —variables, listas, funciones, estructuras condicionales y manejo básico de archivos— para calcular y clasificar un indicador económico realista: el crecimiento trimestral del PIB. Así, se refuerzan habilidades de programación básicas en un contexto económico.

## Paso 1: Definir variables con datos económicos

Supongamos que tenemos datos trimestrales de crecimiento del PIB (en %) de un país.

```python
crecimiento_pib = [1.2, 1.5, 1.3, 1.4]
trimestres = ["T1", "T2", "T3", "T4"]
```

## Paso 2: Calcular el crecimiento promedio

Crea una función que calcule el promedio de crecimiento.

```python
def promedio(lista):
    suma = 0
    for valor in lista:
        suma += valor
    return suma / len(lista)

promedio_crecimiento = promedio(crecimiento_pib)
print(f"Crecimiento promedio anual del PIB: {promedio_crecimiento:.2f}%")
```

## Paso 3: Clasificar el crecimiento

Usa estructuras condicionales para clasificar el crecimiento promedio como: 

- "Alto" (>1.4%)

- "Moderado" (1.2% a 1.4%)

- "Bajo" (<1.2%).

```python
if promedio_crecimiento > 1.4:
    categoria = "Alto"
elif promedio_crecimiento >= 1.2:
    categoria = "Moderado"
else:
    categoria = "Bajo"

print(f"Categoría de crecimiento: {categoria}")
```

## Paso 4: Guardar resultados en un archivo

Guarda el resultado en un archivo de texto llamado `resultado_pib.txt`.

```python
with open("resultado_pib.txt", "w") as archivo:
    archivo.write(f"Crecimiento promedio anual del PIB: {promedio_crecimiento:.2f}%")
    archivo.write(f"Categoría de crecimiento: {categoria}")
```

## Paso 5: Leer y mostrar el archivo guardado

Lee el archivo y muestra su contenido en pantalla.

```python
with open("resultado_pib.txt", "r") as archivo:
    contenido = archivo.read()
    print("Contenido del archivo resultado_pib.txt:")
    print(contenido)
```
