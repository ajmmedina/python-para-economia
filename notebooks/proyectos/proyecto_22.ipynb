# **Proyecto:** Modelado financiero con Python

**Propósito**

Que el estudiante desarrolle modelos financieros simples para evaluar escenarios y tomar decisiones fundamentadas.

## Pasos

1. Definir variables financieras y parámetros.

2. Crear funciones para cálculos financieros (VPN, TIR).

3. Simular escenarios con diferentes parámetros.

4. Analizar resultados y tomar decisiones.

5. Documentar el modelo y resultados.

## Reflexión

- Este proyecto fortalece habilidades para modelar y analizar decisiones financieras con programación.

## Retos adicionales

1. Añadir simulaciones Monte Carlo.

2. Incorporar análisis de sensibilidad.

3. Crear interfaces para entrada de datos.

4. Integrar con bases de datos.

5. Automatizar reportes financieros

## ¿Qué aprendimos con este proyecto?

- Fundamentos de modelado financiero con Python.

- Uso de funciones y simulaciones.

- Análisis y toma de decisiones basadas en modelos.

- Documentación y presentación de resultados.
