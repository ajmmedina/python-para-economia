# **Proyecto:** Análisis de regresión lineal simple con `statsmodels`

**Propósito**

Realizarás un análisis de regresión lineal simple para modelar la relación entre dos variables económicas y evaluar su significancia

## Pasos

1. Importar `pandas` y `statsmodels`.

2. Cargar y preparar datos económicos.

3. Definir variables dependiente e independiente.

4. Ajustar modelo de regresión lineal simple.

5. Interpretar coeficientes y resultados estadísticos.

6. Visualizar ajuste y residuos.

## Reflexión

- Este proyecto introduce técnicas estadísticas para modelar y entender relaciones económicas.

## Retos adicionales

1. Realizar regresión múltiple.

2. Evaluar supuestos del modelo.

3. Aplicar transformaciones a variables.

4. Comparar modelos con diferentes variables.

5. Automatizar reportes de análisis.

## ¿Qué aprendimos con este proyecto?

- Preparación de datos para análisis estadístico.

- Ajuste y evaluación de modelos de regresión.

- Interpretación de coeficientes y pruebas de hipótesis.

- Visualización de resultados y diagnóstico.
