# **Proyecto:** Diseño y uso de clases para análisis económico

**Propósito**

Diseñarás clases que representen modelos económicos, como mercados o agentes, y crearás métodos para simular comportamientos y calcular resultados.

## Pasos

1. Definir clases para agentes económicos con atributos y métodos.

2. Crear clases para mercados que gestionen agentes y transacciones.

3. Implementar métodos para simular intercambios y calcular indicadores.

4. Instanciar objetos y ejecutar simulaciones.

5. Mostrar resultados y analizar comportamientos.

## Reflexión

- Este proyecto muestra cómo la POO permite construir modelos económicos dinámicos y flexibles para análisis y simulación

## Retos adicionales

1.	Añadir herencia para diferentes tipos de agentes.

2.	Implementar eventos aleatorios en simulaciones.

3.	Crear interfaces para interacción con el modelo.

4.	Guardar y cargar estados de simulación.

5.	Visualizar resultados con gráficos.

## ¿Qué aprendimos con este proyecto?

- Diseño de clases complejas y relaciones entre objetos.

- Implementación de métodos para simulación y cálculo.

- Modelado de sistemas económicos con programación.

- Análisis de resultados de simulaciones.
