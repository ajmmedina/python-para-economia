# **Proyecto:** Análisis básico de datos económicos con pandas

**Propósito**
Aprenderás a importar datos económicos reales, limpiarlos y obtener resúmenes estadísticos básicos usando pandas.

## Pasos

1. Importar pandas y cargar un archivo CSV con datos económicos (por ejemplo, inflación mensual).

2. Explorar los datos con funciones como .head(), .info() y .describe().

3. Limpiar datos faltantes o erróneos.

4. Calcular estadísticas básicas (media, mediana, desviación estándar).

5. Guardar un resumen en un nuevo archivo CSV.

## Reflexión

- Este proyecto te permite familiarizarte con pandas para manejar datos económicos reales, una habilidad fundamental para análisis posteriores.

## Retos adicionales

1. Importar datos de diferentes fuentes y formatos.

2. Filtrar datos por fechas o categorías.

3. Crear nuevas columnas con cálculos derivados.

4. Manejar datos faltantes con diferentes estrategias.

5. Comparar estadísticas entre diferentes periodos.


## ¿Qué aprendimos con este proyecto?

- Cómo importar y explorar datos con pandas.

- Técnicas básicas de limpieza de datos.

- Cálculo de estadísticas descriptivas.

- Guardar resultados para análisis futuros.
