# **Proyecto:** Consultas avanzadas en bases de datos con SQL

**Propósito**

Aprenderás a realizar consultas SQL avanzadas, como joins, agrupamientos y subconsultas, para analizar datos económicos.

## Pasos

1. Crear tablas relacionadas, por ejemplo `Indicadores` y `Paises`.

2. Insertar datos en ambas tablas.

3. Realizar consultas con `JOIN` para combinar información.

4. Usar `GROUP BY` para agrupar datos y calcular promedios.

5. Ejecutar subconsultas para filtrar datos específicos.

6. Mostrar resultados y exportar a CSV.

## Reflexión

- Este proyecto fortalece la capacidad para extraer y analizar datos complejos de bases de datos, esencial en economía.

## Retos adicionales

1. Crear vistas para consultas frecuentes.

2. Optimizar consultas con índices.

3. Implementar funciones agregadas personalizadas.

4. Realizar consultas con condiciones temporales.

5. Automatizar reportes con consultas SQL.

## ¿Qué aprendimos con este proyecto?

- Uso de joins para combinar tablas relacionadas.

- Agrupamiento y agregación de datos con SQL.

- Subconsultas para consultas específicas.

- Integración de SQL con Python para análisis.
