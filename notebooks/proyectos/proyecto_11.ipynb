# **Proyecto:** Introducción a la Programación Orientada a Objetos (POO) en Python

**Propósito**

Aprenderás a crear clases y objetos en Python para representar conceptos económicos, como un producto financiero o un cliente bancario.

## Pasos

1. Definir una clase `ProductoFinanciero` con atributos como nombre, tasa de interés y plazo.

2. Crear un método para calcular el interés simple.

3. Instanciar objetos con diferentes productos financieros.

4. Mostrar la información y el interés calculado para cada producto.

5. Crear una clase `Cliente` que tenga una lista de productos financieros y un método para calcular el total de intereses.

## Reflexión

- Este proyecto te permite entender cómo la POO ayuda a organizar y modelar datos complejos en economía, facilitando la reutilización y mantenimiento del código.

## Retos adicionales

1. Añadir métodos para calcular interés compuesto.

2. Crear herencia para diferentes tipos de productos financieros.

3. Implementar validaciones en los atributos.

4. Añadir métodos para mostrar reportes resumidos.

5. Crear una clase `Banco` que gestione múltiples clientes.

## ¿Qué aprendimos con este proyecto?

1. Conceptos básicos de clases, atributos y métodos.

2. Cómo modelar entidades económicas con POO.

3. La importancia de la encapsulación y reutilización en programación.

4. Cómo relacionar objetos para representar estructuras más complejas.
