# Proyecto: Análisis de inflación de Perú

En este proyecto analizaremos datos reales de inflación.

Los economistas usan datos para:

- analizar tendencias
- comparar periodos
- visualizar información

Usaremos Python para analizar una serie simple de inflación anual.

## Paso 1: Importar librerías

Las librerías son herramientas que amplían lo que Python puede hacer.

En este proyecto usaremos:

- pandas → para manejar datos
- matplotlib → para hacer gráficos

### Escribe el siguiente código:

>import pandas as pd

>import matplotlib.pyplot as plt

### Precaución:

Antes de importar librerías debe estar seguro que están instaladas.

Para instalar una librería, existen varias formas, una de ellas es usar el comando `pip`.

## Paso 2: Crear los datos

Usaremos una serie de inflación anual de ejemplo.

### Escribe el siguiente código.

>años = [2018, 2019, 2020, 2021, 2022, 2023]

>inflacion = [1.3, 2.1, 2.0, 6.4, 8.5, 3.2]

## Paso 3: Crear un DataFrame

Un DataFrame es una tabla de datos.

### Escribe el siguiente código:

>datos = pd.DataFrame({

>&emsp;&emsp;"Año": años,

>&emsp;&emsp;"Inflacion": inflacion

>})

## Paso 4: Mostrar los datos

Para ver la tabla usamos:

>print(datos)

## Paso 5: Crear un gráfico

Los economistas usan gráficos para analizar tendencias.

### Escribe el siguiente código:

>plt.plot(datos["Año"], datos["Inflacion"])

>plt.xlabel("Año")

>plt.ylabel("Inflación (%)")

>plt.title("Inflación en Perú")

>plt.show()
