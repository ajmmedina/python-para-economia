# **Proyecto:** Visualización básica de datos económicos con `matplotlib`

**Propósito**

Aprenderás a crear gráficos básicos (líneas, barras) para visualizar datos económicos y entender mejor sus patrones.


# Pasos

1. Importar `matplotlib` y cargar datos económicos simples.

2. Crear un gráfico de línea para mostrar evolución temporal.

3. Crear un gráfico de barras para comparar categorías.

4. Personalizar títulos, etiquetas y leyendas.

5. Guardar los gráficos como imágenes

## Reflexión

- Visualizar datos es clave para comunicar resultados y detectar patrones en economía.

## Retos adicionales

1. Crear gráficos combinados (líneas y barras).

2. Añadir anotaciones y estilos personalizados.

3. Visualizar múltiples series en un solo gráfico.

4. Explorar otros tipos de gráficos (pastel, histograma).

5. Integrar gráficos en reportes automáticos.

## ¿Qué aprendimos con este proyecto?

- Cómo crear gráficos básicos con `matplotlib`.

- Personalización simple de gráficos.

- Guardar gráficos para reportes
