# **Proyecto:** Análisis de series temporales económicas con `pandas`

**Propósito**

Aprenderás a importar datos de series temporales económicas, analizarlos y visualizar tendencias y patrones con gráficos

## Pasos

1. Importar `pandas` y `matplotlib`.

2. Cargar un archivo CSV con datos de series temporales (ejemplo: inflación mensual).

3. Convertir la columna de fechas a formato `datetime`.

4. Realizar análisis descriptivo y cálculo de medias móviles.

5. Crear gráficos de línea para visualizar tendencias.

6. Guardar gráficos como imágenes

## Reflexión

- Este proyecto te permite entender cómo trabajar con datos temporales, fundamentales para análisis económicos y financieros

## Retos adicionales

1. Calcular y graficar tasas de variación mensual y anual.

2. Añadir análisis de estacionalidad.

3. Comparar series temporales de diferentes indicadores.

4. Implementar suavizado exponencial.

5. Exportar análisis a reportes automáticos.

## ¿Qué aprendimos con este proyecto?

-  Importación y manejo de datos temporales con `pandas`.

- Análisis estadístico básico de series temporales.

- Visualización de tendencias con `matplotlib`.

- Interpretación de patrones temporales en economía.
