# **Proyecto 3:** Análisis simple de desempleo trimestral

**Propósito**

Este proyecto introduce el uso de listas, funciones y condicionales para analizar tasas de desempleo trimestrales y clasificarlas.

## Paso 1: Definir listas con datos de desempleo

```python
desempleo = [6.5, 6.7, 6.4, 6.3]
trimestres = ["T1", "T2", "T3", "T4"]
```

## Paso 2: Calcular promedio de desempleo

```python
def promedio(lista):
    return sum(lista) / len(lista)

promedio_desempleo = promedio(desempleo)
print(f"Promedio anual de desempleo: {promedio_desempleo:.2f}%")
```

## Paso 3: Clasificar desempleo

```python
if promedio_desempleo > 7:
    categoria = "Alto"
elif promedio_desempleo >= 5:
    categoria = "Moderado"
else:
    categoria = "Bajo"

print(f"Categoría de desempleo: {categoria}")
```

## Paso 4: Guardar resultados en archivo

```python
with open("resultado_desempleo.txt", "w") as archivo:
    archivo.write(f"Promedio anual de desempleo: {promedio_desempleo:.2f}%")
    archivo.write(f"Categoría de desempleo: {categoria}")
```

## Paso 5: Leer y mostrar archivo

```python
with open("resultado_desempleo.txt", "r") as archivo:
    contenido = archivo.read()
    print("Contenido del archivo resultado_desempleo.txt:")
    print(contenido)
```
